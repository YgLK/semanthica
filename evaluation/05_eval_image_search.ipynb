## Prepare vector collection 

```python
from qdrant_client import QdrantClient
import os 

collection_name = "image_search_fashion"
QDRANT_HOST = os.getenv("QDRANT_DB_HOST")
QDRANT_PORT = os.getenv("QDRANT_DB_PORT")

client = QdrantClient(QDRANT_HOST, port=QDRANT_PORT)
```

```python
[collection.name for collection in client.get_collections().collections]
```

```python
from qdrant_client.http import models

client.recreate_collection(
    collection_name=collection_name,
    vectors_config={
        "image": models.VectorParams(size=2048, distance=models.Distance.DOT),
    },
    on_disk_payload = True
)
```

```python
[collection.name for collection in client.get_collections().collections]
```

## Index items

```python
def generate_payload(row):
    embedding = generate_embedding(row["path"])
    result = client.upsert(
        collection_name=collection_name,
        points=[
            models.PointStruct(
                id=row["id"],
                vector={"image": embedding},
                payload={
                    "item_id": row["id"], 
                    "gender": row["gender"], 
                    "subCategory": row["subCategory"],
                    "articleType": row["articleType"],
                    "image_path": row["path"], 
                }
            )
        ]
    )
    
    if not result:
        print(row)

df_styles.apply(generate_payload, axis=1)
```

```python
collection = client.get_collection(collection_name=collection_name)
collection.points_count
```

```python
def search_by_image(image_path: str, limit=15):
    embedding = generate_embedding(image_path)
    hits = client.search(
        collection_name=collection_name,
        query_vector=models.NamedVector(
            name="image",
            vector=embedding
        ),
        limit=limit,
    )
    return hits


import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def visualize_search_results(query_image_row, hits = 15, num_rows=3, num_items_per_row=5):
    # Load the query image
    query_image = mpimg.imread(query_image_row["path"])

    # Plot results 
    fig, axs = plt.subplots(num_rows, num_items_per_row + 1, figsize=(18, 9))

    # Display the query image
    axs[0, 0].imshow(query_image)
    axs[0, 0].axis('off')
    axs[0, 0].set_title("Query Image")

    # Display search results
    for i in range(num_rows):
        for j in range(num_items_per_row):
            index = i * num_items_per_row + j
            if index < len(hits):
                hit = hits[index]
                img = mpimg.imread(hit.payload["image_path"])
                axs[i, j + 1].imshow(img)
                axs[i, j + 1].axis('off')
                axs[i, j + 1].set_title(f"{hit.payload['articleType']} (score: {hit.score:.2f})")
    
    # put empty image to the rest of the grid
    for i in range(num_rows):
        for j in range(num_items_per_row - len(hits)):
            axs[i, len(hits) + 1 + j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Example usage
query_image_row = df_styles[df_styles['articleType'] == 'Tshirts'].iloc[0]
hits = search_by_image(query_image_row["path"])
visualize_search_results(query_image_row, hits)
```

## Search and evaluate

```python
import pandas as pd
import matplotlib.pyplot as plt

def search_and_evaluate_category(category, top_k=15):
    counter = {
        "category": category,
        "classifications": {cat: 0 for cat in df_styles["articleType"].unique()}
    }

    for idx, row in df_styles[df_styles['articleType'] == category].iterrows():
        hits = search_by_image(row["path"], limit=top_k)
        hits_df = pd.DataFrame([{**hit.payload, "score": hit.score} for hit in hits])
        hits_df["score"] = hits_df["score"].astype(float)
        for hit in hits:
            counter["classifications"][hit.payload["articleType"]] += 1
    
    return counter

counters = []
for category in df_styles["articleType"].unique():
    print(f"Searching for category '{category}'")
    counter = search_and_evaluate_category(category)
    counters.append(counter)
```

```python
counters_df = pd.DataFrame(counters)

counters_df["total"] = counters_df["classifications"].apply(lambda x: sum(x.values()))
for category in df_styles["articleType"].unique():
    counters_df[category] = counters_df.apply(lambda x: x["classifications"][category], axis=1)
```

```python
counters_final = counters_df[["category", "total"] + list(df_styles["articleType"].unique())]
counters_final = counters_final.rename(columns={"category": "query image category"})
```

```python
counters_final.to_csv("image_search_results.csv", index=False)
```
